```python
import pandas as pd
all_data = pd.read_csv("/content/Bidding Mock Analysis - Mock Analysis Data.csv").rename({'Place the bid for your preference 1': 'Bids_pf_1'}, axis=1)

student_limits = {}

for name in all_data['Name']:
  student_limits[name] = 4

course_limits = {
    'Introduction to psychology':18,
    'French':18,
    'Heritage: concepts and practices':18,
    'Consumer Behavior':18,
    'Understanding culture':18,
    'Gender sensitisation':18,
    'Human computer interaction':18,
    'Biomimicry':18
}
pf_1 = all_data.iloc[:,0:3]
num_students = pd.DataFrame(pf_1['Preference 1'].value_counts().reset_index().values, columns=["Course", "Count"])
courses_to_check=[]
courses_with_less = num_students[num_students['Count'] <= 18]['Course']
if len(num_students[num_students['Count'] > 18]) == 0:
  results = pf_1.iloc[:,0:2]
results = pf_1[pf_1['Preference 1'].isin(courses_with_less)].iloc[:,0:2]
courses_to_check = num_students[num_students['Count'] > 18]['Course']

if len(courses_to_check)!=0:
  for course in courses_to_check:
    students = pf_1[pf_1['Preference 1'] == course].sort_values(by='Bids_pf_1', ascending=False).head(18)[:,0:2]
    results = pd.concat([results, students])

num_winners_pf_1 = pd.DataFrame(results['Preference 1'].value_counts().reset_index().values, columns=["Course", "Count"])

for i in range(len(num_winners_pf_1)):
  course_limits[num_winners_pf_1['Course'][i]] -= num_winners_pf_1['Count'][i]

for name in results['Name']:
  student_limits[name] -=1
```
